The data that will be used in this assignemnt is the MAGIC Gamma Telescope Data Set. It can be found here: https://archive.ics.uci.edu/ml/datasets/MAGIC+Gamma+Telescope. This data has 19020 instances and 11 attributes; one of which being the prediction. The data is trying to take in information from a telescope to predict if a signal event is from gamma radiation or just background noise.

```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import pandas as pd
from copy import deepcopy as copy

import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'   #TF2: to fix Jupyter down

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

import IPython.display as ipd  # for display and clear_output
```

The first thing to do is import the data. Here is an explination straight from the page for what each feature is[1]:
1. fLength: continuous # major axis of ellipse [mm]
2. fWidth: continuous # minor axis of ellipse [mm]
3. fSize: continuous # 10-log of sum of content of all pixels [in #phot]
4. fConc: continuous # ratio of sum of two highest pixels over fSize [ratio]
5. fConc1: continuous # ratio of highest pixel over fSize [ratio]
6. fAsym: continuous # distance from highest pixel to center, projected onto major axis [mm]
7. fM3Long: continuous # 3rd root of third moment along major axis [mm]
8. fM3Trans: continuous # 3rd root of third moment along minor axis [mm]
9. fAlpha: continuous # angle of major axis with vector to origin [deg]
10. fDist: continuous # distance from origin to center of ellipse [mm]
11. class: g,h # gamma (signal), hadron (background)

```python
df = pd.read_csv("magic04.csv",names=["fLength","fWidth","fSize","fConc","fConc1","fAsym","fM3Long","fM3Trans","fAlpha","fDist","class"])
df
```

The next thing to do is convert the g and h in the column class into a number the computer can understand.
<br/>g = 0
<br/>h = 1

```python
Tdic = {'g': 0, 'h':1}

df['class'] = df['class'].apply(lambda x: Tdic[x]) 
df
```

Then using pandas' scatter_matrixx we will be able to look for any corrilations between each of the features of the data. Looking at it we can tell that there are distinct paterns in some of the plots.

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, alpha=0.2, figsize=(11,11), diagonal='kde');
```

The next thing to do is look at the data to see if it needs to be standardized or not. From the table below it does need to be standarized.

```python
df.describe()
```

To standardize the data we will abstract a standardize method. Then split up the data into input and target. Finaly, the input data will be standardized.

```python
def standardize(X, mean, std):
    return (X-mean) / std
```

```python
X = df.iloc[:, :-1].copy()
T = df.iloc[:, -1].copy()

mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

XS = standardize(X, mean, std)
XS.describe()
```

Next, lets see if we can find any corralations between the inputs and the target. The only observation I can make from these graphs is that gamma signals usally have a low fWidth and a low fLength.

```python
fig = plt.figure(figsize=(16,12))
plt.clf() # Clear previous plt figure
for i in range(10):
    plt.subplot(3, 4, i+1) # Selects which subplot to plot to
    plt.scatter(X.iloc[:, i], T) # Plots a given column
    plt.ylabel("class") # Sets Y label
    plt.xlabel(X.columns.values[i])
fig.tight_layout()
```

After looking at the data the next step is to partition it for use. We will be importing the partition file[2]. The data will be partitioned into both training and testing data for the inputs and target.

```python
import partitionlib as ptl
# just in case, you made change in the file, you may need to reload as below
import importlib
import partitionlib
importlib.reload(partitionlib);
```

```python
def partition_data(X, T):
    """Partition data into train and test splits."""
    data, targets = ptl.partition(copy(X.values), copy(T.values))
    Xtrain, Xtest = data
    Ttrain, Ttest = targets
    
    return Xtrain, Xtest, Ttrain, Ttest
```

```python
Xtrain, Xtest, Ttrain, Ttest = partition_data(X, T)
print("Train data shape: {}".format(Xtrain.shape))
print("Train target shape: {}".format(Ttrain.shape))
print("Test data shape: {}".format(Xtest.shape))
print("Test target shape: {}".format(Ttest.shape))
```

```python
N, D = X.shape
Ntrain, Ntest = Xtrain.shape[0], Xtest.shape[0]
print("Number of total samples: {}".format(N))
print("Number of total features: {}".format(D))
print("Number of training samples: {}".format(Ntrain))
print("Number of testing samples: {}".format(Ntest))
```

## Control Model

The first model out of the four is a control model. It is just a shallow nn. This model will just be used to see how the dnn function compared to a shallow one.

### Control Structure
Thid model has 10 inputs and 10 hidden units for its hidden layer. The activation function that will be used for this model is a hyperbolic tangent. 

```python
modelControl = keras.Sequential([
    keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(2, activation="softmax")
])

modelControl.summary()
```

```python
modelControl.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
```

### Control Training
The training accuracy for the conrol isn't great but it is also not bad. It got to a 78.99% accuracy. This is a good time to note something about this data set. In the notes for the data it is mentioned that accuracy isn't meaningfull. Insead it is very important that false negitives are kept low. They mention how classifying a background event as a signal is worse than the other way around. 

```python
history = modelControl.fit(Xtrain, Ttrain, epochs=10, validation_split=0.2)
```

With that being said lets take a look at the confusion matrix for the testing data. As we can see the false positives for this model is 996. This gives us a recall of .7557.

```python
from sklearn.metrics import confusion_matrix
Ytrain_probs = modelControl.predict(Xtrain)
Ytrain_preds = np.argmax(Ytrain_probs, axis=1)
confusion_matrix(Ytrain_preds, Ttrain)
```

## Model One

The first model that will be looked at is a basic deep nn. This will also be used to get a good feel for how the other two models might improve upon this basic structure.

### Model One Structure
This model will have 5 hidden layers that are each 10 hidden units. Each of the hidden layers are also using the activation function hyperbolic tangent. The output will have the activation set to softmax and will output 2 numbers.

```python
modelOne = keras.Sequential([
    keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(2, activation="softmax")
])

modelOne.summary()
```

```python
modelOne.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
```

### Model One Training
In comparision to the shallow nn the accuracy has gone down by adding more layers. Lets see if by adding more layers or adjusting the number of hidden units will provide more accuracy. My prediction is that becuase of how many features this data has it will require one more data in general to get great results and two a more layers to really capture the paterns in the data.

```python
history = modelOne.fit(Xtrain, Ttrain, epochs=10, validation_split=0.2)
```

With model one the accuracy is definitly worse but this model is more beneficial by having only 313 false positives. This gives us a recall of .8215.

```python
Ytrain_probs = modelOne.predict(Xtrain)
Ytrain_preds = np.argmax(Ytrain_probs, axis=1)
confusion_matrix(Ytrain_preds, Ttrain)
```

## Model Two

The idea with the second model was to increase how many layers deep this network is. The though is with more layers the more it can capture the paterns in the data.

### Model Two Structure
This model has 10 hidden layers each with 10 hidden units. Each of the layers also still uses the hyperbolic tangent as the activation function.

```python
modelTwo = keras.Sequential([
    keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(D, activation="tanh"),
    tf.keras.layers.Dense(2, activation="softmax")
])

modelTwo.summary()
```

```python
modelTwo.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
```

## Model Two Traning
With the second model the training accuracy has gone up to .74 which is better than the first model but still not as good as the shallow nn. Still getting closer in that regard though. 

```python
historyTwo = modelTwo.fit(Xtrain, Ttrain, epochs=10, validation_split=0.2)
```

Unforutently the model is predicting more false postives but that might be becuase its predicting more positives in general. The recall of model two is .7994. This is better than the shallow nn but barely worse than model one.

```python
Ytrain_probs = modelTwo.predict(Xtrain)
Ytrain_preds = np.argmax(Ytrain_probs, axis=1)
confusion_matrix(Ytrain_preds, Ttrain)
```

## Model Three

The idea with the third model was to scale it back to 5 layers like model one but try increasing the number of hidden units.

### Model Three Structure
The structure of model three is 5 hidden layers with 20 hidden units each. Double the number of inputs. Again each layer but the output layer is using an activation of hyperbolic tangent. It is also important to note this model by far has the most params.

```python
modelThree = keras.Sequential([
    keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(D*2, activation="tanh"),
    tf.keras.layers.Dense(D*2, activation="tanh"),
    tf.keras.layers.Dense(D*2, activation="tanh"),
    tf.keras.layers.Dense(D*2, activation="tanh"),
    tf.keras.layers.Dense(D*2, activation="tanh"),
    tf.keras.layers.Dense(2, activation="softmax")
])

modelThree.summary()
```

```python
modelThree.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
```

### Model Three Training

The accuracy for this model was the best out of the deep nn, but none of the models were able to beat the shallow nn accuracy.

```python
historyThree = modelThree.fit(Xtrain, Ttrain, epochs=10, validation_split=0.2)
```

Looking at the false positives it has the highest out of all of the models. The recall is .7279. This is much lower than the first models recall.

```python
Ytrain_probs = modelThree.predict(Xtrain)
Ytrain_preds = np.argmax(Ytrain_probs, axis=1)
confusion_matrix(Ytrain_preds, Ttrain)
```

## Training Thoughts

So far it seems like model one does the best with recall which from what the data says is the most important analytic. Model one also doesn't have as many parameters as model two and three so it should use less resources. Lets see if this all holds up true with the testing data.
